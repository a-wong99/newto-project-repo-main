### 2. Taking Orders
Now that we have our menu, let's allow customers to place orders. We'll use a `while` loop to let them order multiple items until they are done, and store their selections in a list.
#### Demo Example: Simple Order Taking

```python
# Initialize an empty list to store customer orders
customer_order = []

while True:
    order_item = input("What would you like to order? (Type 'done' to finish): ").strip().title()
    if order_item == 'Done':
        break
    elif order_item in menu:
        customer_order.append(order_item)
        print(f"Added {order_item} to your order.")
    else:
        print("Sorry, that item is not on our menu. Please choose from the list.")

print("Your current order: ", customer_order)
```

```python
import datetime
import random
```

```python
def display_menu(menu):
    """Displays the coffee shop menu with neatly aligned items and prices."""
    print("\n--- Our Menu ---")
    for item, price in menu.items():
        print(f"{item.ljust(16)}: £{price:.2f}")
    print("----------------")
```

```python
def take_order(menu):
    """Allows a customer to place an order, including quantity, with input validation."""
    customer_name = input("Hello! What's your name? ").strip().title()
    current_order = []

    print(f"\nWelcome, {customer_name}! Please place your order.")

    while True:
        order_item = input("What would you like to order? (Type 'Done' to finish): ").strip().title()

        if order_item == 'Done':
            break
        elif order_item in menu:
            while True:
                try:
                    order_quantity_str = input(f"How many {order_item}s would you like? ").strip()
                    order_quantity = int(order_quantity_str)
                    if order_quantity > 0:
                        break
                    else:
                        print("Quantity must be a positive number. Please try again.")
                except ValueError:
                    print("Invalid input. Please enter a whole number for the quantity.")

            current_order.append((order_item, order_quantity))
            if order_quantity == 1:
                print(f"Added {order_item} to your order.")
            else:
                print(f"Added {order_quantity} {order_item}s to your order.")
        else:
            print("Sorry, that item is not on our menu. Please choose from the list.")

    print("\nYour order is: ", current_order)
    return customer_name, current_order
```

```python
def calculate_bill(customer_order_items, menu, VAT_RATE=0.20):
    """Calculates the subtotal, VAT, service charge, and total bill for an order."""
    subtotal = 0
    for item, quantity in customer_order_items:
        subtotal += menu[item] * quantity

    # Apply service charge if subtotal is over £15
    service_charge_amount = 0
    if subtotal > 15:
        SERVICE_CHARGE_RATE = 0.10 # 10% service charge
        service_charge_amount = subtotal * SERVICE_CHARGE_RATE

    tax_amount = subtotal * VAT_RATE
    total_before_discount = subtotal + tax_amount

    return subtotal, tax_amount, service_charge_amount, total_before_discount
```

```python
def apply_discounts_and_lucky_draw(total_bill_before_discount):
    """Applies tiered discounts and conducts a conditional lucky draw."""
    discount_amount = 0
    total_bill_after_discount = total_bill_before_discount
    discount_percentage = 0

    if total_bill_before_discount > 15:
        discount_percentage = 0.15
        discount_amount = total_bill_before_discount * discount_percentage
        total_bill_after_discount -= discount_amount
        print(f"Congratulations! You received a {discount_percentage*100:.0f}% discount of £{discount_amount:.2f}.")
    elif total_bill_before_discount > 10:
        discount_percentage = 0.10
        discount_amount = total_bill_before_discount * discount_percentage
        total_bill_after_discount -= discount_amount
        print(f"Congratulations! You received a {discount_percentage*100:.0f}% discount of £{discount_amount:.2f}.")
    else:
        print("No discount applied.")

    lucky_prize_message = "Better luck next time in our Lucky Draw!"
    if total_bill_after_discount > 5: # Only get a chance if total bill is above £5
        lucky_prizes = ["free cookie", "free coffee", "10% off next order"]
        if random.random() < 0.01: # 1% chance for Grand Prize
            lucky_prize_message = '*** Lucky Draw! You won our Grand Prize! Enjoy a free meal for your next visit! ***'
        elif random.random() < 0.3: # 30% chance for regular prize
            prize = random.choice(lucky_prizes)
            lucky_prize_message = f'*** Lucky Draw! You won a {prize}! ***'
    else:
        lucky_prize_message = "See you next time!"

    print(lucky_prize_message)
    return discount_amount, total_bill_after_discount, lucky_prize_message
```

```python
def generate_receipt(customer_data, menu):
    """Generates and prints a receipt, and saves it to a file."""
    name = customer_data['customer_name']
    order_items = customer_data['customer_order']
    subtotal = customer_data['subtotal']
    tax_amount = customer_data['tax_amount']
    service_charge = customer_data['service_charge_amount']
    discount_amount = customer_data['discount_amount']
    final_total = customer_data['final_total']
    current_date = customer_data['date']

    receipt_content = []
    receipt_content.append("--- Your Coffee Shop Receipt ---")
    receipt_content.append(f"Name: {name}")
    receipt_content.append(f"Date: {current_date}")
    receipt_content.append("Item            Qty   Price   Total")
    receipt_content.append("-----------------------------------")

    for item, quantity in order_items:
        item_price = menu[item]
        line_total = item_price * quantity
        receipt_content.append(f"{item:<15} {quantity:<5} £{item_price:<7.2f} £{line_total:.2f}")

    receipt_content.append("-----------------------------------")
    receipt_content.append(f"Subtotal:        £{subtotal:>8.2f}")
    receipt_content.append(f"VAT:             £{tax_amount:>8.2f}")
    if service_charge > 0:
        receipt_content.append(f"Service Charge:  £{service_charge:>8.2f}")
    if discount_amount > 0:
        receipt_content.append(f"Discount:        -£{discount_amount:>7.2f}")
    receipt_content.append(f"Total Due:       £{final_total:>8.2f}")
    receipt_content.append("-----------------------------------")
    receipt_content.append("Thank you for your order!")
    receipt_content.append(customer_data['lucky_prize_message'])
    receipt_content.append("\n") # Add an extra newline for separation in the file

    # Print to console
    for line in receipt_content:
        print(line)

    # Save to file
    filename = f"receipt_{name.replace(' ', '_')}_{current_date}.txt"
    with open(filename, 'w') as f:
        for line in receipt_content:
            f.write(line + '\n')
    print(f"Receipt saved to {filename}")
```

```python
all_transactions = []

# Menu defined globally as per original setup
menu = {
    'Coffee': 2.50,
    'Tea': 2.00,
    'Croissant': 3.00,
    'Muffin': 2.75,
    'Hot Chocolate': 3.25,
    'Chocolate Cake': 4.20
}

VAT_RATE = 0.20 # Define VAT rate globally for consistency

while True:
    display_menu(menu)
    customer_name, customer_order_items = take_order(menu)

    if not customer_order_items:
        print(f"\n{customer_name}, your order is empty. Moving to the next customer or ending session.")
        continue # Skip to next iteration or exit if no items

    subtotal, tax_amount, service_charge_amount, total_before_discount = calculate_bill(customer_order_items, menu, VAT_RATE)

    discount_amount, final_total_after_discount, lucky_prize_message = apply_discounts_and_lucky_draw(total_before_discount)

    transaction_data = {
        'customer_name': customer_name,
        'date': datetime.date.today(),
        'customer_order': customer_order_items,
        'subtotal': subtotal,
        'tax_amount': tax_amount,
        'service_charge_amount': service_charge_amount,
        'total_before_discount': total_before_discount,
        'discount_amount': discount_amount,
        'final_total': final_total_after_discount,
        'lucky_prize_message': lucky_prize_message
    }
    all_transactions.append(transaction_data)

    print("\nGenerating receipt...")
    generate_receipt(transaction_data, menu)

    another_customer = input("\nProcess another customer? (yes/no): ").strip().lower()
    if another_customer != 'yes':
        break

print("\n--- End of Day Summary ---")
print(f"Total customers served: {len(all_transactions)}")
overall_revenue = sum(t['final_total'] for t in all_transactions)
print(f"Overall revenue: £{overall_revenue:.2f}")
print("\nAll transactions details:")
for i, transaction in enumerate(all_transactions):
    print(f"\nCustomer {i+1}:")
    for key, value in transaction.items():
        if isinstance(value, (float, int)) and 'amount' in key or 'total' in key or 'subtotal' in key:
            print(f"  {key}: £{value:.2f}")
        else:
            print(f"  {key}: {value}")

print("\nCoffee shop system closed. Have a great day!")
```
